```python
#elbow method to find clusters for a data set. any way we know there are 3 classes from target labels. lets conform that.
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=2)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

#from the plot optimum numbers of clusters we can say those are 3 clusters.
```

```python
# from this plots we can take "plot 6" and "plot 2" for prediction
#     iris.data[:, 0] - sepal length in cm
#     iris.data[:, 1] - sepal width in cm
#     iris.data[:, 2] - petal length in cm
#     iris.data[:, 3] - petal width in cm

x=iris.data[:, 2];
y1=iris.data[:, 0]
y2=iris.data[:, 3]
```

```python
km = KMeans(n_clusters=3)
km.fit(X)
predict=km.predict(X)
print(predict)
```

```python
label = {0: 'red', 1: 'blue', 2: 'green'}
plt.scatter(x, y1, c=predict)
plt.title("Petal Length vs Sepal Length")
plt.show()
```

```python
label = {0: 'red', 1: 'blue', 2: 'green'}
plt.scatter(x, y2, c=predict)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(iris.target, predict, target_names=['setosa', 'versicolor', 'virginica']))
```
